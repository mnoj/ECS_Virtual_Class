# Resonance Filter Theory

## 1. Basic Parameters

A **resonance filter** is a frequency-selective network that passes a specific band of frequencies while attenuating others. The performance and selectivity of these filters are characterized by several key parameters.

- The **Circuit Q** represents the ratio of the center frequency $f_c$ of the resonant circuit to its bandwidth (BW). A higher Q indicates a narrower, more selective filter, and $f_2$ and $f_1$ are the upper and lower 3-dB frequencies.

$$
Q = \frac{f_c}{BW} = \frac{f_2 - f_1}{f_c}
$$

- The **Shape Factor (SF)** is the ratio of the 60-dB bandwidth to the 3-dB bandwidth of the resonant circuit. It describes the "steepness" of the filter's skirts; a shape factor closer to 1 implies an ideal, vertical transition between the passband and stopband.

$$
SF = \frac{BW_{3dB}}{BW_{60dB}}
$$

- **Ripple** is a measure of the flatness of the passband of a resonant circuit. It quantifies the variations in signal amplitude within the desired frequency range.

$$
\text{Ripple (dB)} = A_{\text{max}} - A_{\text{min}} \quad \text{(within passband)}
$$

- **Insertion Loss (IL)** is the attenuation caused by the components inserted between a generator and its load. This represents the power lost simply by adding the filter to the circuit.

$$
IL(dB) = 10 \log_{10} \left( \frac{P_{\text{load, with filter}}}{P_{\text{load, without filter}}} \right)
$$


## 2. Resonance of a Parallel LC Circuit

In a **parallel LC circuit**, resonance occurs when the inductive reactance $$X_L$$ and the capacitive reactance $X_C$ are equal in magnitude but opposite in phase. At resonance, the total reactance of the circuit becomes zero, and the impedance reaches its maximum value. The resonance condition for a parallel LC circuit is given by:

$$
f_{\text{res}} = \frac{1}{2 \pi \sqrt{LC}}
$$

Where:
- **$f_{\text{res}}$** is the resonant frequency (in Hz),
- **$L$** is the inductance (in Henrys),
- **$C$** is the capacitance (in Farads).

## Loaded $Q$ of a Circuit with an Ideal Capacitor and a Non-Ideal Inductor

In real circuits, the inductor is not ideal, and it exhibits a finite  $Q_L$ factor. The **loaded $Q$** of a circuit is a measure of the damping or losses in the circuit, and it changes when the circuit is loaded with a resistive component or when the inductor itself has losses. 



# Basic Filters

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import FloatLogSlider, HBox, VBox, interactive_output, Label
from IPython.display import display

def update_plot(R_lpf, C_nf, R_hpf, L_uh):
    f = np.logspace(2, 8, 1000)
    w = 2 * np.pi * f
    
    # LPF (RC) Math
    C_f = C_nf * 1e-9
    h_lpf = (1/(1j*w*C_f)) / (R_lpf + 1/(1j*w*C_f))
    mag_lpf = 20 * np.log10(np.abs(h_lpf))
    fc_lpf = 1 / (2 * np.pi * R_lpf * C_f)
    
    # HPF (RL) Math
    L_h = L_uh * 1e-6
    h_hpf = (1j*w*L_h) / (R_hpf + 1j*w*L_h)
    mag_hpf = 20 * np.log10(np.abs(h_hpf))
    fc_hpf = R_hpf / (2 * np.pi * L_h)

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 7))
    
    # --- Helper for Frequency Formatting ---
    def fmt_f(val):
        return f"{val/1e3:.2f} kHz" if val >= 1e3 else f"{val:.0f} Hz"

    # --- Plot LPF (Octave with Coordinates) ---
    ax1.semilogx(f, mag_lpf, color='blue', lw=2)
    ax1.axvline(fc_lpf, color='red', linestyle='--', alpha=0.3)
    
    # Points: P1 at fc, P2 at 2*fc
    f2_lpf = fc_lpf * 2
    ax1.plot([fc_lpf, f2_lpf], [-3, -9], 'o--', color='orange', lw=2, markersize=8)
    
    # Coordinate Labels
    ax1.text(fc_lpf, 2, f"({fmt_f(fc_lpf)}, -3dB)", color='blue', fontweight='bold', ha='right')
    ax1.text(f2_lpf, -15, f"({fmt_f(f2_lpf)}, -9dB)", color='orange', fontweight='bold', ha='left')
    
    ax1.set_title('LPF: 6 dB Octave Drop')
    ax1.set_ylim(-50, 10)
    ax1.grid(True, which="both", alpha=0.3)

    # --- Plot HPF (Octave with Coordinates) ---
    ax2.semilogx(f, mag_hpf, color='green', lw=2)
    ax2.axvline(fc_hpf, color='red', linestyle='--', alpha=0.3)
    
    # Points: P1 at fc/2, P2 at fc
    f1_hpf = fc_hpf / 2
    ax2.plot([f1_hpf, fc_hpf], [-9, -3], 'o--', color='orange', lw=2, markersize=8)
    
    # Coordinate Labels
    ax2.text(f1_hpf, -15, f"({fmt_f(f1_hpf)}, -9dB)", color='orange', fontweight='bold', ha='right')
    ax2.text(fc_hpf, 2, f"({fmt_f(fc_hpf)}, -3dB)", color='green', fontweight='bold', ha='left')
    
    ax2.set_title('HPF: 6 dB Octave Rise')
    ax2.set_ylim(-50, 10)
    ax2.grid(True, which="both", alpha=0.3)
    
    plt.tight_layout()
    plt.show()

# Standard Sliders
s_r_lpf = FloatLogSlider(value=1000, base=10, min=1, max=5, description='R (Ω)')
s_c_lpf = FloatLogSlider(value=10, base=10, min=0, max=4, description='C (nF)')
s_r_hpf = FloatLogSlider(value=50, base=10, min=1, max=4, description='R (Ω)')
s_l_hpf = FloatLogSlider(value=10, base=10, min=-1, max=4, description='L (µH)')

controls = HBox([VBox([Label('LPF Octave View'), s_r_lpf, s_c_lpf]), 
                 VBox([Label('HPF Octave View'), s_r_hpf, s_l_hpf])])
out = interactive_output(update_plot, {'R_lpf': s_r_lpf, 'C_nf': s_c_lpf, 'R_hpf': s_r_hpf, 'L_uh': s_l_hpf})

display(controls, out)
```

# Resonant Filters

### Ideal Resonator

```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display

%matplotlib inline

def updateFilter(l_nh, c_pf):
    # 1. Component Conversions
    L = l_nh * 1e-9    # Inductor in nH
    C = c_pf * 1e-12   # Capacitor in pF
    rSource = 50
    rLoad = 50
    # 2. Natural Resonance Calculus
    # fc = 1 / (2 * pi * sqrt(LC))
    f_res = 1 / (2 * np.pi * np.sqrt(L * C))
    
    # Characteristic Impedance Zo = sqrt(L/C)
    zo = np.sqrt(L / C)
    
    # Loaded Q (Ideal tank): QL = R_parallel / Zo
    rParallel = (rSource * rLoad) / (rSource + rLoad)
    qLoaded = rParallel / zo
    
    # Resulting Bandwidth: BW = fc / QL
    bwActual = f_res / qLoaded

    # 3. Simulation Range (centered on resonance)
    freqs = np.logspace(np.log10(f_res/50), np.log10(f_res*50), 5000)
    omega = 2 * np.pi * freqs
    
    # Ideal Tank Impedance
    zTank = 1 / ((1/(1j * omega * L)) + (1j * omega * C))
    
    # Gain Calculation (Vout/Vin relative to no-filter case)
    zParallel = 1 / (1/zTank + 1/rLoad)
    vRatio = zParallel / (rSource + zParallel)
    vRef = rLoad / (rSource + rLoad)
    gainDB = 20 * np.log10(np.abs(vRatio / vRef))

    # 4. Plotting
    fig, ax = plt.subplots(figsize=(12, 7))
    ax.semilogx(freqs/1e6, gainDB, color='blue', lw=2, label='Ideal LC Response')
    
    def label_pt(f, db, col, offset, label_txt=None):
        ax.plot(f/1e6, db, 'o', color=col)
        txt = label_txt if label_txt else f"({f/1e6:.1f}M, {db:.1f}dB)"
        ax.text(f/1e6, db + offset, txt, color=col, ha='center', fontsize=9, fontweight='bold')

    # Resonance Point
    label_pt(f_res, 0, 'red', 2, f"fc: {f_res/1e6:.1f} MHz")
    
    # 12 dB/octave Verification (One octave up)
    f_oct_near = f_res * 2
    label_pt(f_oct_near, -12, 'red', -4)
    ax.plot([f_res/1e6, f_oct_near/1e6], [0, -12], '--', color='red', label='12 dB/oct')

    # 6 dB/octave Verification (At the high frequency limit)
    f_far = freqs[-300]
    idx1 = np.argmin(np.abs(freqs - f_far))
    db1 = gainDB[idx1]
    f_far_oct = f_far * 2
    label_pt(f_far, db1, 'orange', 2)
    label_pt(f_far_oct, db1-6, 'orange', -4)
    ax.plot([f_far/1e6, f_far_oct/1e6], [db1, db1-6], '--', color='orange', label='6 dB/oct')

    ax.set_title(f"Natural Resonance | L={l_nh:.1f}nH, C={c_pf:.1f}pF\nfc={f_res/1e6:.2f}MHz")
    ax.set_xlabel("Frequency (MHz)")
    ax.set_ylabel("Relative Gain (dB)")
    ax.set_ylim(-70, 15)
    ax.grid(True, which='both', alpha=0.3)
    ax.legend(loc='upper right')
    plt.show()

# Widgets
lS = widgets.FloatSlider(value=20.0, min=10, max=50, step=0.01, description='L (nH)')
cS = widgets.FloatSlider(value=200.0,min=150, max=300, description='C (pF)')

ui = widgets.VBox([widgets.HBox([lS, cS])])
out = widgets.interactive_output(updateFilter, {'l_nh': lS, 'c_pf': cS})
display(ui, out)
```

### Resonator Design with Ideal Lumped Elements (Variable Rs)

```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display, Markdown

%matplotlib inline

def updateFilter(rSource):
    # 1. FIXED Component Values
    l_uh = 0.05
    c_pf = 25.0
    L = l_uh * 1e-6    
    C = c_pf * 1e-12   
    
    # 2. Natural Resonance Calculus (Static)
    f_res = 1 / (2 * np.pi * np.sqrt(L * C))
    zo = np.sqrt(L / C) 
    
    # 3. Dynamic Loaded Q Calculation
    # With no RL, Rp is simply Rs
    r_parallel_total = rSource
    # QL = Rs / Zo
    q_loaded = r_parallel_total / zo
    # BW = fc / QL
    bw_3db = f_res / q_loaded
    
    # 4. Simulation Range
    freqs = np.logspace(np.log10(f_res/50), np.log10(f_res*50), 5000)
    omega = 2 * np.pi * freqs
    
    # Ideal Tank Impedance
    zTank = 1 / ((1/(1j * omega * L)) + (1j * omega * C))
    
    # Gain Calculation (Vout is voltage across the tank)
    # vRatio = zTank / (rSource + zTank)
    # Since there is no RL, vRef is 1 (direct connection)
    vRatio = zTank / (rSource + zTank)
    gainDB = 20 * np.log10(np.abs(vRatio))

    # 5. Plotting
    fig, ax = plt.subplots(figsize=(12, 7))
    ax.semilogx(freqs/1e6, gainDB, color='blue', lw=2, label=f'QL = {q_loaded:.2f}')
    
    def label_pt(f, db, col, offset, label_txt=None):
        ax.plot(f/1e6, db, 'o', color=col)
        txt = label_txt if label_txt else f"({f/1e6:.1f}M, {db:.1f}dB)"
        ax.text(f/1e6, db + offset, txt, color=col, ha='center', fontsize=9, fontweight='bold')

    # Center Frequency Marker
    label_pt(f_res, 0, 'red', 2, f"fc: {f_res/1e6:.1f} MHz")
    
    # -3dB Bandwidth Markers
    ax.axhline(-3, color='green', linestyle=':', alpha=0.5, label='-3 dB Level')

    ax.set_title(f"Impact of Rs on Q (L={l_uh}uH, C={c_pf}pF)")
    ax.set_xlabel("Frequency (MHz)")
    ax.set_ylabel("Response (dB)")
    ax.set_ylim(-60, 10)
    ax.grid(True, which='both', alpha=0.3)
    ax.legend(loc='upper right')
    plt.show()

    # Results Display
    display(Markdown(f"**Loaded Q ($Q_L$): {q_loaded:.2f}**"))

# Widget: Only Rs is adjustable
rsS = widgets.FloatSlider(value=450.0, min=10, max=5000, step=10, description='Rs (Ω)')

ui = widgets.VBox([rsS])
out = widgets.interactive_output(updateFilter, {'rSource': rsS})
display(ui, out)
```

### Resonator Design with Ideal Lumped Elements

```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display, Markdown

%matplotlib inline

def updateFilter(rSource, rLoad, fCenterMhz, bwMhz):
    # 1. Physical Constants
    fCenter = fCenterMhz * 1e6      
    bwTarget = bwMhz * 1e6
    qLoadedTarget = fCenter / bwTarget
    
    # Resistance seen by the tank (External resistances only)
    rTotal = (rSource * rLoad) / (rSource + rLoad)
    
    # 2. Reactance and Component Values
    xpVal = rTotal / qLoadedTarget
    lVal = xpVal / (2 * np.pi * fCenter)
    cVal = 1 / (2 * np.pi * fCenter * xpVal)

    # 3. Simulation
    f_min = max(1e5, fCenter - 5 * bwTarget)
    f_max = fCenter + 5 * bwTarget
    freqs = np.linspace(f_min, f_max, 5000)
    omega = 2 * np.pi * freqs
    
    zTank = 1 / ((1/(1j * omega * lVal)) + (1j * omega * cVal))
    zParallel = 1 / (1/zTank + 1/rLoad)
    vRatio = zParallel / (rSource + zParallel)
    vRef = rLoad / (rSource + rLoad)
    
    gainDB = 20 * np.log10(np.abs(vRatio / vRef))
    peakGain = 0 

    # 4. Reference Circuit (Constant Q = 1.1)
    qRef = 1.1
    zo_ref = rTotal / qRef
    lRef = zo_ref / (2 * np.pi * fCenter)
    cRef = 1 / (2 * np.pi * fCenter * zo_ref)
    
    zTankRef = 1 / ((1/(1j * omega * lRef)) + (1j * omega * cRef))
    zParallelRef = 1 / (1/zTankRef + 1/rLoad)
    vRatioRef = zParallelRef / (rSource + zParallelRef)
    gainDBRef = 20 * np.log10(np.abs(vRatioRef / vRef))

    # 5. Plotting
    fig, ax = plt.subplots(figsize=(12, 7))
    
    # Static Reference
    ax.plot(freqs/1e6, gainDBRef, color='gray', lw=1.5, linestyle='--', label='Reference (Q=1.1)', alpha=0.5)
    
    # Interactive Main Tank
    ax.plot(freqs/1e6, gainDB, color='blue', lw=2.5, label=f'Main Tank (Q={qLoadedTarget:.2f})')
    
    # 6 dB/octave Asymptotic Marker
    f_high_oct = fCenter * 2
    if f_high_oct < f_max:
        idx_oct = np.argmin(np.abs(freqs - f_high_oct))
        db_oct = gainDB[idx_oct]
        ax.plot(f_high_oct/1e6, db_oct, 'o', color='orange')
        ax.text(f_high_oct/1e6, db_oct + 2, f"({f_high_oct/1e6:.1f}M, {db_oct:.1f}dB)", 
                color='orange', ha='center', fontweight='bold')

    # Formatting - Added numerical value of fc to the legend label
    ax.axvline(fCenter/1e6, color='red', alpha=0.3, linestyle=':', label=f'fc = {fCenterMhz:.1f} MHz')
    ax.set_title(f"Ideal Filter Design: Main Tank vs. Q=1.1 Reference")
    ax.set_xlabel("Frequency (MHz)")
    ax.set_ylabel("Response (dB)")
    ax.set_ylim(-50, 5)
    ax.grid(True, which='both', alpha=0.2)
    ax.legend(loc='upper right')
    
    plt.show()

    # Design values displayed at the bottom
    display(Markdown(f"Design Values: L = **{lVal*1e9:.2f} nH** | C = **{cVal*1e12:.2f} pF**"))
    display(Markdown(f"Calculated Parameters: Xp = **{xpVal:.2f} Ω** | Rp = **{rTotal:.2f} Ω** | Loaded Q = **{qLoadedTarget:.2f}**"))

# Widget UI
fcS = widgets.FloatSlider(value=50.0, min=25.0, max=150.0, step=1.0, description='fc (MHz)')
bwS = widgets.FloatSlider(value=2.5, min=1.0, max=10, step=0.1, description='BW (MHz)')
rsS = widgets.FloatSlider(value=150.0, min=10.0, max=500.0, step=10.0, description='Rs (Ω)')
rlS = widgets.FloatSlider(value=1500.0, min=10.0, max=2000.0, step=10.0, description='Rl (Ω)')

ui = widgets.VBox([widgets.HBox([fcS, bwS]), widgets.HBox([rsS, rlS])])
out = widgets.interactive_output(updateFilter, {
    'fCenterMhz': fcS, 'bwMhz': bwS, 'rSource': rsS, 'rLoad': rlS
})

display(ui, out)
```

### Resonator Design with Non-Ideal Lumped Elements

```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display, Markdown

%matplotlib inline

def updateFilter(rSource, rLoad, fCenterMhz, bwMhz, qInductor):
    # 1. Physical Constants
    fCenter = fCenterMhz * 1e6      
    bwTarget = bwMhz * 1e6
    qLoadedTarget = fCenter / bwTarget
    
    # Resistance seen by the tank (External resistances only)
    rExternal = (rSource * rLoad) / (rSource + rLoad)
    
    # 2. Reactance and Component Values
    # We calculate L and C based on the target Loaded Q assuming ideal components
    xpVal = rExternal / qLoadedTarget
    lVal = xpVal / (2 * np.pi * fCenter)
    cVal = 1 / (2 * np.pi * fCenter * xpVal)
    
    # 3. Non-Ideal Effects (Inductor Q)
    # The inductor losses appear as a parallel resistor
    rParallelInd = qInductor * xpVal
    # The new actual total resistance including inductor losses
    rTotal = 1 / (1/rExternal + 1/rParallelInd)
    # The actual Loaded Q of the real circuit
    qLoadedActual = rTotal / xpVal

    # 4. Simulation
    f_min = max(1e5, fCenter - 5 * bwTarget)
    f_max = fCenter + 5 * bwTarget
    freqs = np.linspace(f_min, f_max, 5000)
    omega = 2 * np.pi * freqs
    
    # Tank impedance including the inductor loss resistor
    zTank = 1 / ((1/(1j * omega * lVal)) + (1j * omega * cVal) + (1/rParallelInd))
    zParallel = 1 / (1/zTank + 1/rLoad)
    vRatio = zParallel / (rSource + zParallel)
    
    # Reference (Vout/Vin with no filter)
    vRef = rLoad / (rSource + rLoad)
    
    gainDB = 20 * np.log10(np.abs(vRatio / vRef))
    peakGain = np.max(gainDB)

    # 5. Reference Circuit (Constant Q = 1.1)
    qRef = 1.1
    zo_ref = rExternal / qRef
    lRef = zo_ref / (2 * np.pi * fCenter)
    cRef = 1 / (2 * np.pi * fCenter * zo_ref)
    zTankRef = 1 / ((1/(1j * omega * lRef)) + (1j * omega * cRef))
    zParallelRef = 1 / (1/zTankRef + 1/rLoad)
    vRatioRef = zParallelRef / (rSource + zParallelRef)
    gainDBRef = 20 * np.log10(np.abs(vRatioRef / vRef))

    # 6. Plotting
    fig, ax = plt.subplots(figsize=(12, 7))
    
    ax.plot(freqs/1e6, gainDBRef, color='gray', lw=1.5, linestyle='--', label='Reference (Q=1.1)', alpha=0.5)
    ax.plot(freqs/1e6, gainDB, color='blue', lw=2.5, label=f'Main Tank (Actual Q={qLoadedActual:.2f})')
    
    f_high_oct = fCenter * 2
    if f_high_oct < f_max:
        idx_oct = np.argmin(np.abs(freqs - f_high_oct))
        db_oct = gainDB[idx_oct]
        ax.plot(f_high_oct/1e6, db_oct, 'o', color='orange')
        ax.text(f_high_oct/1e6, db_oct + 2, f"({f_high_oct/1e6:.1f}M, {db_oct:.1f}dB)", 
                color='orange', ha='center', fontweight='bold')

    ax.axvline(fCenter/1e6, color='red', alpha=0.3, linestyle=':', label=f'fc = {fCenterMhz:.1f} MHz')
    ax.set_title(f"Filter Design with Inductor Losses: Peak Gain = {peakGain:.2f} dB")
    ax.set_xlabel("Frequency (MHz)")
    ax.set_ylabel("Response (dB)")
    ax.set_ylim(peakGain - 50, peakGain + 5)
    ax.grid(True, which='both', alpha=0.2)
    ax.legend(loc='upper right')
    
    plt.show()

    display(Markdown(f"Design Values: L = **{lVal*1e9:.2f} nH** | C = **{cVal*1e12:.2f} pF**"))
    display(Markdown(f"Calculated Parameters: Xp = **{xpVal:.2f} Ω** | Rp (External) = **{rExternal:.2f} Ω** | **Actual Loaded Q = {qLoadedActual:.2f}**"))

# Widget UI
fcS = widgets.FloatSlider(value=50.0, min=25.0, max=150.0, step=1.0, description='fc (MHz)')
bwS = widgets.FloatSlider(value=2.5, min=1.0, max=10, step=0.1, description='BW (MHz)')
qS = widgets.FloatSlider(value=100.0, min=10.0, max=1000.0, step=1.0, description='Ind Q')
rsS = widgets.FloatSlider(value=1000.0, min=10.0, max=500.0, step=10.0, description='Rs (Ω)')
rlS = widgets.FloatSlider(value=1500.0, min=10.0, max=2000.0, step=10.0, description='Rl (Ω)')

ui = widgets.VBox([widgets.HBox([fcS, bwS, qS]), widgets.HBox([rsS, rlS])])
out = widgets.interactive_output(updateFilter, {
    'fCenterMhz': fcS, 'bwMhz': bwS, 'rSource': rsS, 'rLoad': rlS, 'qInductor': qS
})

display(ui, out)
```

# 3. Exercise

```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
import schemdraw
import schemdraw.elements as elm
from IPython.display import display, Markdown

# Use %matplotlib inline for standard Jupyter notebooks
%matplotlib inline

def updateTappedC(f0Mhz, bwMhz, rSource, rLoad, qInd):
    # Clear previous plots
    plt.close('all')
    
    # 1. Input Conversions
    f0 = f0Mhz * 1e6
    bw = bwMhz * 1e6
    w0 = 2 * np.pi * f0
    
    # 2. Design Calculus
    qe = f0 / bw
    if qInd <= qe:
        display(Markdown("## Error: Inductor Q must be higher than Loaded Q (Qe)!"))
        return

    rp = rLoad * (qInd - qe) / (2 * qe)
    xp = rp / qInd 
    lp = xp / w0
    cTotal = 1 / (xp * w0)
    k = np.sqrt(rLoad / rSource) - 1
    c2 = cTotal * (1 + k) / k
    c1 = k * c2

    # 3. Simulation Logic
    freqMin, freqMax = f0 - (5 * bw), f0 + (5 * bw)
    freqs = np.linspace(freqMin, freqMax, 2000)
    omega = 2 * np.pi * freqs
    yTank = (1 / (1j * omega * lp)) + (1 / rp) + (1 / rLoad)
    zTankBranch = 1 / yTank
    zSeriesBranch = (1 / (1j * omega * c2)) + zTankBranch
    yInput = (1j * omega * c1) + (1 / zSeriesBranch)
    zInputTotal = 1 / yInput
    vRatioWithFilter = (zInputTotal / (rSource + zInputTotal)) * (zTankBranch / zSeriesBranch)
    vPeakIdeal = 0.5 * np.sqrt(rLoad / rSource) 
    gainDB = 20 * np.log10(np.abs(vRatioWithFilter / vPeakIdeal))
    ilAtResonance = np.max(gainDB)

    # 4. Plotting
    fig, ax = plt.subplots(figsize=(10, 4))
    ax.plot(freqs/1e6, gainDB, color='royalblue', lw=2)
    ax.axvline(f0/1e6, color='red', linestyle='--', alpha=0.5)
    ax.axvspan((f0 - bw/2)/1e6, (f0 + bw/2)/1e6, color='green', alpha=0.1, label='3dB BW')
    ax.axhline(ilAtResonance - 3, color='orange', linestyle=':', label='-3 dB Level')
    ax.set_title(f"Response: fc={f0Mhz:.2f}MHz, BW={bwMhz:.2f}MHz, IL={ilAtResonance:.2f}dB")
    ax.set_xlabel("Frequency (MHz)")
    ax.set_ylabel("Insertion Loss (dB)")
    ax.set_xlim(freqMin/1e6, freqMax/1e6)
    ax.set_ylim(ilAtResonance - 50, 2)
    ax.grid(True, which='both', alpha=0.3)
    ax.legend(loc='upper right')
    plt.show()

    # 5. Professional Schematic Rendering
    display(Markdown("### Circuit Topology: Tapped-C Matching Network"))
    
    d = schemdraw.Drawing()
    d.config(unit=2, fontsize=10)
    
    # Source and Rs
    d += (In := elm.Dot().label('Vin', 'left'))
    d += (RS := elm.Resistor().right().label(f'Rs\n{rSource:.0f}Ω', 'top'))
    d += elm.Dot()
    
    # C1 Shunt - Centered label with offset
    d.push()
    d += elm.Capacitor().down().label(f'C1: {c1*1e12:.1f}pF', loc='right', rotate=90, ofst=0.5)
    d += elm.Ground()
    d.pop()
    
    # C2 Series
    d += (C2 := elm.Capacitor().right().label(f'C2\n{c2*1e12:.1f}pF', 'top'))
    d += (Dot2 := elm.Dot())
    
    # Tank Section - Centered labels with offset
    d.push()
    d += elm.Inductor().down().label(f'Lp: {lp*1e9:.1f}nH', loc='right', rotate=90, ofst=0.5)
    d.pop()
    
    d += elm.Line().right().at(Dot2.end).length(1.5)
    d.push()
    d += elm.Resistor().down().label(f'Rp: {rp:.0f}Ω', loc='right', rotate=90, ofst=0.5)
    d.pop()
    
    d += elm.Line().right().length(1.5)
    d += elm.Resistor().down().label(f'Rl: {rLoad:.0f}Ω', loc='right', rotate=90, ofst=0.5)
    
    # Bottom Rail Connection
    d += elm.Line().left().tox(Dot2.end)
    d += elm.Line().left().tox(RS.end)
    
    display(d.draw())

    # 6. Data Summary
    display(Markdown(rf"""
**Design Parameters:**
* **Reactance $X_p$:** {xp:.2f} $\Omega$ | **Inductor Loss $R_p$:** {rp:.2f} $\Omega$
* **Inductor $L_p$:** {lp*1e9:.2f} nH | **Match Factor $k$:** {k:.2f}
"""))

# Slider Setup
fcS = widgets.FloatSlider(value=100.0, min=1.0, max=500.0, step=1.0, description='fc (MHz)')
bwS = widgets.FloatSlider(value=5.0, min=0.1, max=50.0, step=0.1, description='BW (MHz)')
qS = widgets.FloatSlider(value=100.0, min=10.0, max=500.0, step=5.0, description='Ind Q')
rsS = widgets.FloatSlider(value=50.0, min=10.0, max=1000.0, step=10.0, description='Rs (Src)')
rlS = widgets.FloatSlider(value=2000.0, min=100.0, max=10000.0, step=100.0, description='Rl (Ld)')

ui = widgets.VBox([widgets.HBox([fcS, bwS]), widgets.HBox([qS, rsS, rlS])])
out = widgets.interactive_output(updateTappedC, {
    'f0Mhz': fcS, 'bwMhz': bwS, 'qInd': qS, 'rSource': rsS, 'rLoad': rlS
})

display(ui, out)
```

# 4. Cascaded Tapped C

```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display, Markdown
import schemdraw
import schemdraw.elements as elm

%matplotlib inline

def updateCascadedTappedC(f0Mhz, bwMhz, rSource, rLoad, qInd):
    plt.close('all')
    
    # 1. Input Conversions
    f0 = f0Mhz * 1e6
    bw = bwMhz * 1e6
    w0 = 2 * np.pi * f0
    
    # 2. Design Calculus
    # Increase stage Qe by sqrt(2) for cascading correction
    qeSingle = (f0 / bw) * np.sqrt(2)
    
    if qInd <= qeSingle:
        display(Markdown("## Error: Inductor Q must be higher than Stage Qe!"))
        return

    # Rp calculation: rp = rLoad * (qInd - qeSingle) / qeSingle
    rp = rLoad * (qInd - qeSingle) / qeSingle
    xp = rp / qInd 
    lp = xp / w0
    cTotal = 1 / (xp * w0)
    
    kMatch = np.sqrt(rLoad / rSource) - 1
    c2 = cTotal * (1 + kMatch) / kMatch
    c1 = kMatch * c2
    c12 = cTotal / qeSingle
    
    # The tuning capacitor for the second tank
    ct = cTotal



    # 4. Simulation
    freqMin, freqMax = f0 - (5 * bw), f0 + (5 * bw)
    freqs = np.linspace(freqMin, freqMax, 3000)
    omega = 2 * np.pi * freqs
    
    s = 1j * omega
    kCoupling = 1 / qeSingle
    transfer = (kCoupling * s * w0) / (s**2 + (w0/qeSingle)*s + w0**2)
    
    gainDB = 10 * np.log10(np.abs(transfer)**2 / np.max(np.abs(transfer)**2))

    # 5. Plotting
    fig, ax = plt.subplots(figsize=(10, 4))
    ax.plot(freqs/1e6, gainDB, color='purple', lw=2)
    ax.axvspan((f0 - bw/2)/1e6, (f0 + bw/2)/1e6, color='green', alpha=0.1)
    ax.set_title(f"Cascaded Response: fc={f0Mhz:.2f}MHz, Stage Qe={qeSingle:.2f}")
    ax.set_ylabel("Gain (dB)")
    ax.set_xlabel("Frequency (MHz)")
    ax.grid(True, alpha=0.3)
    plt.show()

    display(Markdown(f"""
**Design Results (2 Decimals):**
* **Tuning Cap $C_t$:** {ct*1e12:.2f} pF
* **Coupling $C_{{12}}$:** {c12*1e12:.2f} pF
* **Inductance $L_p$:** {lp*1e9:.2f} nH
* **Shunt $C_1$:** {c1*1e12:.2f} pF
* **Series $C_2$:** {c2*1e12:.2f} pF
"""))

    # 3. Schematic Drawing (C2 before Lp1, added Ct at output)
    d = schemdraw.Drawing()
    
    # Stage 1
    d += (V1 := elm.SourceV().label('Vin'))
    d += (RS := elm.Resistor().right().label('Rs'))
    d += elm.Line().dot()
    
    d.push()
    d += elm.Capacitor().down().label('C1', loc='bottom')
    d += elm.Line().left().tox(V1.start)
    d.pop()
    
    d += (C2S1 := elm.Capacitor().right().label('C2'))
    d += elm.Line().dot()
    
    d.push()
    d += elm.Inductor().down().label('Lp1', loc='bottom')
    d.pop()
    
    # Coupling
    d += (C12 := elm.Capacitor().right().label('C12'))
    d += elm.Line().dot()
    
    # Stage 2 Tank (Lp2 || Ct || Rl)
    d.push()
    d += elm.Inductor().down().label('Lp2', loc='bottom')
    d.pop()
    
    d += elm.Line().right().length(1.5).dot()
    d.push()
    d += elm.Capacitor().down().label('Ct', loc='bottom')
    d.pop()
    
    d += elm.Line().right().length(1.5)
    d += elm.Resistor().down().label('Rl', loc='bottom')
    d += elm.Line().left().tox(V1.start)
    
    display(d.draw())

# Slider Setup
fcS = widgets.FloatSlider(value=75.0, min=10, max=500, description='fc')
bwS = widgets.FloatSlider(value=3.75, min=0.1, max=20,step=0.01, description='BW')
qS = widgets.FloatSlider(value=85, min=50, max=500, description='Ind Q')
rsS = widgets.FloatSlider(value=100, min=10, max=500, description='Rs')
rlS = widgets.FloatSlider(value=1000, min=100, max=5000, description='Rl')

ui = widgets.VBox([widgets.HBox([fcS, bwS]), widgets.HBox([qS, rsS, rlS])])
out = widgets.interactive_output(updateCascadedTappedC, {'f0Mhz': fcS, 'bwMhz': bwS, 'qInd': qS, 'rSource': rsS, 'rLoad': rlS})
display(ui, out)
```

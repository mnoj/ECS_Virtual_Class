# Analyzing the AMPS Wideband Digital Receiver

In this notebook, we will model the performance of a wideband digital receiver for the AMPS (Advanced Mobile Phone System) standard using the AD6645 14-bit ADC. We will explore how conversion gain, noise figure, and ADC specifications impact system sensitivity and Spurious-Free Dynamic Range (SFDR). The goal of this system is to digitize a $12.5\text{ MHz}$ wide IF (Intermediate Frequency) band containing up to 416 individual channels.

## Part 1: Signal and Sampling Parameters
First, define the core parameters of our system based on the AMPS receiver specifications.

| Parameter            | Symbol      | Value          | Definition                                                                                   |
|----------------------|-------------|----------------|----------------------------------------------------------------------------------------------|
| Sampling Rate        | $f_s$       | 61.44 MSPS     | Chosen as a power-of-two multiple of the channel bandwidth.                                  |
| Nyquist Bandwidth    | $BW_{N}$    | 30.72 MHz      | The effective bandwidth of the first Nyquist zone ($f_s / 2$).                               |
| Channel Bandwidth    | $BW_{ch}$   | 30 kHz         | Default width of a single AMPS communication channel.                                            |
| Full-Scale           | $FS$        | +5 dBm         | The maximum input power level before ADC clipping.                                           |

What is the process gain of our ADC?


```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

# Fixed System Constant
SAMPLING_RATE = 61.44e6       # 61.44 MSPS
ADC_FS_DBM = 5.0              # Full-scale input of ADC in dBm
ADC_SFRD = 96                 # dBc
ADC_NOISE = 3                 # dB ADC internal noise is equal to the noise at the input

def plot_process_gain(bw_khz):
    # Setup the range for the background curve (10kHz to 2000kHz)
    bw_range_khz = np.linspace(10, 2000, 400)
    pg_curve = 10 * np.log10(SAMPLING_RATE / (2 * bw_range_khz * 1000))
    
    # Calculate current point
    current_pg = 10 * np.log10(SAMPLING_RATE / (2 * bw_khz * 1000))
    
    # Create the Plot
    plt.figure(figsize=(10, 6))
    plt.plot(bw_range_khz, pg_curve, label='Process Gain Curve', color='#1f77b4', linewidth=2)
    plt.scatter(bw_khz, current_pg, color='red', s=100, zorder=5, label=f'Current: {current_pg:.2f} dB')
    
    # Formatting
    plt.title(f"Process Gain vs. Channel Bandwidth\n(Sampling Rate = {SAMPLING_RATE/1e6} MSPS)", fontsize=14)
    plt.xlabel("Channel Bandwidth (kHz)", fontsize=12)
    plt.ylabel("Process Gain (dB)", fontsize=12)
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.legend()
    
    # Annotation for the current value
    plt.annotate(f"{current_pg:.2f} dB", 
                 xy=(bw_khz, current_pg), 
                 xytext=(bw_khz + 100, current_pg + 2),
                 arrowprops=dict(facecolor='black', shrink=0.05, width=1, headwidth=5))
    
    plt.show()

    # Data output with carrier returns
    print(f"--- Results ---\n")
    print(f"Selected Bandwidth: {bw_khz} kHz\n")
    print(f"Calculated Process Gain: {current_pg:.2f} dB\n")
    print(f"----------------")

# Create the interactive slider
interact(plot_process_gain, 
         bw_khz=FloatSlider(value=30.0, min=10.0, max=2000.0, step=10.0, 
                            description='BW (kHz):', continuous_update=True));
```
